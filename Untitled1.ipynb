```python

import numpy as np
import torch
from scipy.integrate import odeint
import torch.nn as nn
import matplotlib.pyplot as plt
from PIL import Image
```

```python

def save_gif_PIL(outfile, files, fps=5, loop=0):
    "Helper function for saving GIFs"
    imgs = [Image.open(file) for file in files]
    imgs[0].save(fp=outfile, format='GIF', append_images=imgs[1:], save_all=True, duration=int(1000/fps), loop=loop)
    
def solve_duffing(d, a, b, gamma, w, x):
    def duffing(y, t):
        y0, y1 = y
        dydt = [y1, -d * y1 - a * y0 - b * y0**3 + gamma * np.cos(w * t)]
        return dydt

    y0 = [0, 0]  
    sol = odeint(duffing, y0, x.view(-1).numpy())  # Convert x to a one-dimensional numpy array
    y = torch.tensor(sol[:, 0])
    return y

class FCN(nn.Module):
    def __init__(self, N_INPUT, N_OUTPUT, N_HIDDEN, N_LAYERS):
        super().__init__()
        activation = nn.Tanh
        self.fcs = nn.Sequential(*[
                        nn.Linear(N_INPUT, N_HIDDEN),
                        activation()])
        self.fch = nn.Sequential(*[
                        nn.Sequential(*[
                            nn.Linear(N_HIDDEN, N_HIDDEN),
                            activation()]) for _ in range(N_LAYERS-1)])
        self.fce = nn.Linear(N_HIDDEN, N_OUTPUT)
        
    def forward(self, x):
        x = self.fcs(x)
        x = self.fch(x)
        x = self.fce(x)
        return x

d = 0.001  
a = 0.5 
b = -1.0  
gamma = 0.5
w = 20
x = torch.linspace(0, 1, 500).view(-1,1) 
y = solve_duffing(d, a, gamma, b, w, x).view(-1,1)

x_data = x[0:200:20]
y_data = y[0:200:20]

plt.figure()
plt.plot(x, y, label="Exact solution")
plt.scatter(x_data, y_data, color="tab:orange", label="Training data")
plt.legend()
plt.show()

x_physics = torch.linspace(0, 1, 30).view(-1, 1).requires_grad_(True)

torch.manual_seed(123)
model = FCN(1,1,48,6)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)

files = []

for i in range(50000):
    optimizer.zero_grad()
    yh = model(x_data)
    loss1 = torch.mean((yh - y_data)**2)

    yhp = model(x_physics)
    dy_pred = torch.autograd.grad(yhp, x_physics, torch.ones_like(yhp), create_graph=True)[0]
    dy_pred = dy_pred.requires_grad_(True)
    d2y_pred = torch.autograd.grad(dy_pred, x_physics, torch.ones_like(dy_pred), create_graph=True, allow_unused=True)[0]

    physics = d2y_pred + d * dy_pred + a * yhp + b * torch.pow(yhp, 3) - gamma * torch.cos(w * x_physics)
    loss_physics = (1e-4) * torch.mean(physics** 2)
    total_loss = loss1 + loss_physics

    total_loss.backward()
    optimizer.step()

    if (i+1) % 150 == 0: 
        yh = model(x).detach()
        xp = x_physics.detach()
        
        plot_result(x,y,x_data,y_data,yh,xp)
        
        file = "plots/pinn_%.8i.png"%(i+1)
        #plt.savefig(file, bbox_inches='tight', pad_inches=0.1, dpi=100, facecolor="white")
        #files.append(file)
        
        if (i+1) % 6000 == 0: plt.show()
        else: plt.close("all")
            
#save_gif_PIL("pinn.gif", files, fps=20, loop=0)
# Initialize lists to store displacement and velocity data
displacement_data = []
velocity_data = []

for i in range(20000):
    optimizer.zero_grad()
    yh = model(x_data)
    loss1 = torch.mean((yh - y_data)**2)

    yhp = model(x_physics)
    dy_pred = torch.autograd.grad(yhp, x_physics, torch.ones_like(yhp), create_graph=True)[0]
    dy_pred = dy_pred.requires_grad_(True)
    d2y_pred = torch.autograd.grad(dy_pred, x_physics, torch.ones_like(dy_pred), create_graph=True, allow_unused=True)[0]

    physics = d2y_pred + d * dy_pred + a * yhp + b * torch.pow(yhp, 3) - gamma * torch.cos(w * x_physics)
    loss_physics = (1e-4) * torch.mean(physics** 2)
    total_loss = loss1 + loss_physics

    total_loss.backward()
    optimizer.step()

    # Store displacement and velocity data
    displacement_data.append(yhp.detach().numpy())
    velocity_data.append(dy_pred.detach().numpy())
    
    if (i+1) % 150 == 0: 
        yh = model(x).detach()
        xp = x_physics.detach()
        
        plot_result(x,y,x_data,y_data,yh,xp)
        
        file = "plots/pinn_%.8i.png"%(i+1)
        #plt.savefig(file, bbox_inches='tight', pad_inches=0.1, dpi=100, facecolor="white")
        files.append(file)
        
        if (i+1) % 6000 == 0: plt.show()
        else: plt.close("all")
            
#save_gif_PIL("pinn.gif", files, fps=20, loop=0)
```
